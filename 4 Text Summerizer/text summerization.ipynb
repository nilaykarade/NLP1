```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
from heapq import nlargest
```

```python
STOP_WORDS
```

```python
type(STOP_WORDS)
```

```python
stopwords=list(STOP_WORDS)
print(stopwords)
type(stopwords)
```

```python
!python -m spacy download en_core_web_sm
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
type(nlp)
```

```python
doc=nlp(text)
```

```python
type(doc)
```

```python
tokens=[token.text for token in doc]
```

```python
for token in doc:
    print(token.text)
```

```python
tokens
```

```python
word_frequency={}
```

```python
doc.sents
```

```python
for word in doc:
    if word.text.lower() not in stopwords and word.text not in punctuation:
        if word.text not in word_frequency.keys():
            word_frequency[word.text]=1
        else:
            word_frequency[word.text]+=1
            
#print(word_frequency)
max_frequency=max(word_frequency.values())
            
for word in word_frequency.keys():
    word_frequency[word]=word_frequency[word]/max_frequency

#print(word_frequency)

sent_tokens=[sent for sent in doc.sents]
sent_scores={}

for sent in sent_tokens:
    for word in sent:
        if word.text in word_frequency.keys():
            if word.text in word_frequency.keys():
                sent_scores[sent]=word_frequency[word.text]
            else:
                sent_scores[sent]+=word_frequency[word.text]
#print(sent_scores)

select_len=int(len(sent_tokens)*0.2)

summary=nlargest(select_len,sent_scores,key=sent_scores.get)

#print(summary)


x=[word.text for word in summary]
final_summary=' '.join(x)
final_summary
```
